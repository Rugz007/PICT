# Spliting Data as Train and Test

```python
import sklearn;
from sklearn.model_selection import train_test_split
```

```python
X=df['fheight'].values[:,None]
X.shape
```

```python
y=df.iloc[:,2].values
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
lm.fit(X_train,y_train)
```

```python
y_test=lm.predict(X_test)
print(y_test)
```

# Plotting the Linear Regression Line

```python
plt.figure(figsize=(10, 8))
plt.grid(True, linewidth= 1, linestyle="--")
plt.scatter(X,y,color='orange')
plt.plot(X_test,y_test,color='blue',linewidth=3)
plt.xlabel('Father height in inches')
plt.ylabel('Son height in inches')
plt.show()
```

# Model Performance

```python
y_train_pred=lm.predict(X_train).ravel()
y_test_pred=lm.predict(X_test).ravel()
```

```python
from sklearn.metrics import mean_squared_error as mse,r2_score
```

```python
print("The Mean Squared Error on Train set is:{:0.4f}".format(mse(y_train,y_train_pred)))
print("The Mean Squared Error on Test set is:{:0.4f}".format(mse(y_test,y_test_pred)))
```

```python
print("The R2 score on the Train set is:{:0.4f}".format(r2_score(y_train,y_train_pred)))
print("The R2 score on the Test set is:{:0.4f}".format(r2_score(y_test,y_test_pred)))
```
